```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Number of random points in the donuts
num_points = 3000

# Outer donut parameters
outer_radius_large = 8

# Inner donut parameters
inner_radius_large = 5

# Gap between the donuts
gap_size = 0.5

# Generate random radii and angles for the outer donut
radii_outer = np.random.uniform(inner_radius_large + gap_size, outer_radius_large, num_points)
theta_outer = np.random.uniform(0, 2 * np.pi, num_points)

# Parametric equations for the outer donut shape
x_outer = radii_outer * np.cos(theta_outer)
y_outer = radii_outer * np.sin(theta_outer)

# Inner donut parameters
outer_radius_small = 5

# Generate random radii and angles for the inner donut
radii_inner = np.random.uniform(inner_radius, outer_radius_small - gap_size, num_points)
theta_inner = np.random.uniform(0, 2 * np.pi, num_points)

# Parametric equations for the inner donut shape
x_inner = radii_inner * np.cos(theta_inner)
y_inner = radii_inner * np.sin(theta_inner)

# Plotting the donuts with randomly scattered points and a gap
plt.scatter(x_outer, y_outer, s=5, label='Outer Donut')  # s parameter controls the size of the points
plt.scatter(x_inner, y_inner, s=5, label='Inner Donut')
plt.title('Donuts with Randomly Scattered Points and a Gap')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.axis('equal')  # Equal scaling for a proper visualization
plt.legend()
plt.show()
```

```python
# Create a DataFrame
df_outer = pd.DataFrame({'x': x_outer, 'y': y_outer, 'label': 'Outer Donut'})
df_inner = pd.DataFrame({'x': x_inner, 'y': y_inner, 'label': 'Inner Donut'})

# Concatenate DataFrames
df = pd.concat([df_outer, df_inner])
```

```python
df['label'] = df['label'].replace({'Outer Donut': 1, 'Inner Donut': 0})
```

```python
def Square1(row):
    return float(row['x']**2)

df['x^2'] = df.apply(Square1, axis=1)
df.head()
```

```python
def Square2(row):
    return float(row['y']**2)

df['y^2'] = df.apply(Square2, axis=1)
df.head()
```

```python
def Multiply(row):
    return float(row['y']*row['x'])

df['x*y'] = df.apply(Multiply, axis=1)
df.head()
```

```python
df = df[['x', 'y', 'x^2', 'y^2', 'x*y', 'label']]
df.head()
```
